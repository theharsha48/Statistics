````python
import numpy as np
from scipy import stats
from statistics import median, mode

# Generate a list of 100 integers between 90 and 130
int_list = np.random.randint(90, 131, 100)

# (i) Calculate the mean
def calculate_mean(numbers):
    return sum(numbers) / len(numbers)

print("Mean:", calculate_mean(int_list))

# (i) Calculate the median
def calculate_median(numbers):
    return median(numbers)

print("Median:", calculate_median(int_list))

# (ii) Calculate the mode
def calculate_mode(numbers):
    return mode(numbers)

print("Mode:", calculate_mode(int_list))

# (iii) Calculate the weighted mean
weights = np.random.rand(100)  # random weights
def calculate_weighted_mean(values, weights):
    return np.average(values, weights=weights)

print("Weighted Mean:", calculate_weighted_mean(int_list, weights))

# (iv) Calculate the geometric mean
def calculate_geometric_mean(numbers):
    return stats.gmean(numbers)

print("Geometric Mean:", calculate_geometric_mean(int_list))

# (v) Calculate the harmonic mean
def calculate_harmonic_mean(numbers):
    return len(numbers) / sum(1/x for x in numbers)

print("Harmonic Mean:", calculate_harmonic_mean(int_list))

# (vi) Calculate the midrange
def calculate_midrange(numbers):
    return (min(numbers) + max(numbers)) / 2

print("Midrange:", calculate_midrange(int_list))

# (vii) Calculate the trimmed mean
def calculate_trimmed_mean(numbers, percentage):
    trimmed_numbers = sorted(numbers)[int(percentage*len(numbers)/100):int((100-percentage)*len(numbers)/100)]
    return sum(trimmed_numbers) / len(trimmed_numbers)

print("Trimmed Mean (10%):", calculate_trimmed_mean(int_list, 10))
````

3.Write a Python class representing a discrete random variable with methods to calculate its expected 
value and variance.

````python
class DiscreteRandomVariable:
    def __init__(self, values, probabilities):
        """
        Initialize a discrete random variable.

        Parameters:
        values (list): List of possible values of the random variable.
        probabilities (list): List of probabilities corresponding to each value.
        """
        self.values = values
        self.probabilities = probabilities

    def expected_value(self):
        """
        Calculate the expected value of the discrete random variable.

        Returns:
        float: Expected value of the random variable.
        """
        return sum(x * p for x, p in zip(self.values, self.probabilities))

    def variance(self):
        """
        Calculate the variance of the discrete random variable.

        Returns:
        float: Variance of the random variable.
        """
        ev = self.expected_value()
        return sum((x - ev) ** 2 * p for x, p in zip(self.values, self.probabilities))

values = [1, 2, 3, 4, 5]
probabilities = [0.1, 0.2, 0.3, 0.2, 0.2]
rv = DiscreteRandomVariable(values, probabilities)
print("Expected Value:", rv.expected_value())
print("Variance:", rv.variance())
````

9. Create a program to calculate the cumulative distribution function (CDF) of exponential distribution.

````python
import numpy as np
import scipy.stats as stats

def exponential_cdf(x, rate):
    """
    Calculate the cumulative distribution function (CDF) of an exponential distribution.

    Parameters:
    x (float): The value at which to evaluate the CDF.
    rate (float): The rate parameter of the exponential distribution.

    Returns:
    float: The cumulative probability at x.
    """
    return 1 - np.exp(-rate * x)
rate = 2
x = 1.5
cdf = exponential_cdf(x, rate)
print(f"The cumulative probability at x={x} is {cdf:.4f}")
````

11. A company wants to test if a new website layout leads to a higher conversion rate (percentage of visitors 
who make a purchase). They collect data from the old and new layouts to compare.  
 To generate the data use the following command:  
 ```python  
 import numpy as np   
 # 50 purchases out of 1000 visitors  
 old_layout = np.array([1] * 50 + [0] * 950)  
 # 70 purchases out of 1000 visitors    
new_layout = np.array([1] * 70 + [0] * 930)  
 ```  
 Apply z-test to find which layout is successful.  

````python
from scipy.stats import norm

# Generate the data
old_layout = np.array([1] * 50 + [0] * 950)
new_layout = np.array([1] * 70 + [0] * 930)

# Calculate the sample proportions
p_old = np.mean(old_layout)
p_new = np.mean(new_layout)

# Calculate the pooled proportion
p_pool = (np.sum(old_layout) + np.sum(new_layout)) / (len(old_layout) + len(new_layout))

# Calculate the standard error
se = np.sqrt(p_pool * (1 - p_pool) * (1/len(old_layout) + 1/len(new_layout)))

# Calculate the z-score
z = (p_new - p_old) / se

# Calculate the p-value
p_value = 2 * (1 - norm.cdf(abs(z)))

print(f"z-score: {z:.4f}")
print(f"p-value: {p_value:.4f}")

if p_value < 0.05:
    print("The new layout is significantly better than the old layout.")
else:
    print("There is no significant difference between the two layouts.")
````

 Use the following data:  
 ```python

````python
from scipy.stats import ttest_ind

def ab_test_analysis(clicks_a, clicks_b):
    """
    Perform an A/B test analysis to compare click-through rates of two website layouts.

    Parameters:
    clicks_a (list): Clicks for layout A
    clicks_b (list): Clicks for layout B

    Returns:
    t_stat (float): t-statistic
    df (int): Degrees of freedom
    p_value (float): p-value
    """
    # Calculate the sample means
    mean_a = np.mean(clicks_a)
    mean_b = np.mean(clicks_b)

    # Perform the t-test
    t_stat, p_value = ttest_ind(clicks_a, clicks_b)

    # Calculate the degrees of freedom
    df = len(clicks_a) + len(clicks_b) - 2

    return t_stat, df, p_value

layout_a_clicks = [28, 32, 33, 29, 31, 34, 30, 35, 36, 37]
layout_b_clicks = [40, 41, 38, 42, 39, 44, 43, 41, 45, 47]

t_stat, df, p_value = ab_test_analysis(layout_a_clicks, layout_b_clicks)

print(f"t-statistic: {t_stat:.4f}")
print(f"Degrees of freedom: {df}")
print(f"p-value: {p_value:.4f}")

if p_value < 0.05:
    print("Layout B has a significantly higher click-through rate than Layout A.")
else:
    print("There is no significant difference in click-through rates between Layout A and Layout B.")
````

21. A political analyst wants to determine if there is a significant association between age groups and voter 
preferences (Candidate A or Candidate B). They collect data from a sample of 500 voters and classify 
them into different age groups and candidate preferences. Perform a Chi-Square test to determine if 
there is a significant association between age groups and voter preferences.   
 Use the below code to generate data:  
 ```python  
 np.random.seed(0)  
 age_groups = np.random.choice(['18-30', '31-50', '51+', '51+'], size=30)  
 voter_preferences = np.random.choice(['Candidate A', 'Candidate B'], size=30)  

````python
import pandas as pd
from scipy.stats import chi2_contingency

np.random.seed(0)
age_groups = np.random.choice(['18-30', '31-50', '51+'], size=500)
voter_preferences = np.random.choice(['Candidate A', 'Candidate B'], size=500)

# Create a contingency table
contingency_table = pd.crosstab(age_groups, voter_preferences)

print("Contingency Table:")
print(contingency_table)

# Perform Chi-Square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

print(f"Chi-Square statistic: {chi2:.4f}")
print(f"p-value: {p:.4f}")

if p < 0.05:
    print("There is a statistically significant association between age groups and voter preferences.")
else:
    print("There is no statistically significant association between age groups and voter preferences.")
````

 24. A company produces three different versions of a product: Standard, Premium, and Deluxe. The 
company wants to determine if there is a significant difference in customer satisfaction scores among the 
three product versions. They conducted a survey and collected customer satisfaction scores for each 
version from a random sample of customers. Perform an ANOVA test to determine if there is a significant 
difference in customer satisfaction scores.  
 Use the following data:  
 ```python  
 # Sample data: Customer satisfaction scores for each product version  
 standard_scores = [80, 85, 90, 78, 88, 82, 92, 78, 85, 87]  
 ```  
 premium_scores = [90, 92, 88, 92, 95, 91, 96, 93, 89, 93]  
 deluxe_scores = [95, 98, 92, 97, 96, 94, 98, 97, 92, 99]  

````python
from scipy.stats import f_oneway

# Sample data: Customer satisfaction scores for each product version
standard_scores = [80, 85, 90, 78, 88, 82, 92, 78, 85, 87]
premium_scores = [90, 92, 88, 92, 95, 91, 96, 93, 89, 93]
deluxe_scores = [95, 98, 92, 97, 96, 94, 98, 97, 92, 99]

# Perform ANOVA test
f_stat, p_val = f_oneway(standard_scores, premium_scores, deluxe_scores)

print("F-statistic: {:.4f}".format(f_stat))
print("p-value: {:.4f}".format(p_val))

if p_val < 0.05:
    print("There is a statistically significant difference in customer satisfaction scores among the three product versions.")
else:
    print("There is no statistically significant difference in customer satisfaction scores among the three product versions.")
````
